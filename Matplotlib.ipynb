```python
data = pd.read_csv(url)
```

```python
data.head()
```

## Create a pie chart presenting the male/female proportion 

```python
data['sex'].value_counts()
```

```python
male,female = (data['sex'].value_counts()/len(data))*100
```

```python
print(male)
print(female)
```

```python
labels = 'female', 'male'
explode = (0, 0.1)
```

```python
plt.figure(figsize=(10,10))
plt.pie([male,female],labels=labels,explode=explode, autopct='%1.1f%%' ,shadow=True, startangle=90)
plt.show()
```

## Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x = 'fare',y = 'age',hue='sex',data = data)
```
